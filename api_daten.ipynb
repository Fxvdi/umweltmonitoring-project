```python
import pandas as pd
import requests
import geopandas as gpd
```

```python
def extract_data(response):
    _id = response['_id']
    extracted_data = []
    sensors = response.get('sensors', [])
    for sensor in sensors:
        if sensor.get('title') == 'PM10':
            last_measurement = sensor.get('lastMeasurement', {})
            if last_measurement:
                value = last_measurement.get('value')
                created_at = last_measurement.get('createdAt')
                extracted_data.append({
                    '_id': _id,
                    'created_at': created_at,
                    'value': value
                })
    return extracted_data
```

```python
data = gpd.read_file("data/geodaten.geo.json")
_id = data.id.loc[0]
url = f"https://api.opensensemap.org/boxes/{_id}/sensors"
daten = requests.get(url).json()
extracted_data = extract_data(daten)
df = pd.DataFrame(extracted_data)
print(df)
```

```python
data = gpd.read_file("data/geodaten.geo.json")

dfs = []
for i in range(0, 100):
    _id = data.id.loc[i]
    url = f"https://api.opensensemap.org/boxes/{_id}/sensors"
    daten = requests.get(url).json()
    df = extract_data(daten)
    dfs.extend(df)

# Erstellen eines DataFrames aus der Liste von Dictionaries
df_combined = pd.DataFrame(dfs)
df_combined
```
