```python
import pandas as pd
import plotly.express as px
```

```python
landpm25 = pd.read_csv("werte_daten/werte_land_PM2_5.csv")
landpm10 = pd.read_csv("werte_daten/werte_land_PM10.csv")

stadtpm25 = pd.read_csv("werte_daten/werte_stadt_PM2_5.csv")
stadtpm10 = pd.read_csv("werte_daten/werte_stadt_PM10.csv")

landpm10

```

```python
landpm10['datum'] = pd.to_datetime(landpm10['createdAt'])
columns_to_process = ['value_1', 'value_2', 'value_3', 'value_4', 'value_5']

# Konvertiere die Spalten zu float, falls sie nicht bereits numerisch sind
landpm10[columns_to_process] = landpm10[columns_to_process].apply(pd.to_numeric)

# Interpolieren und dann gruppieren
for col in columns_to_process:
    landpm10[f'{col}_int'] = landpm10[col].interpolate()

# Gruppieren nach 'datum' und Berechnung des Mittelwerts fÃ¼r interpolierte Spalten
# Hier mÃ¼ssen wir sicherstellen, dass wir die numerischen Typen erhalten
grouped_station1 = landpm10.groupby('datum').mean(numeric_only=True)

# Nur die interpolierten Spalten auswÃ¤hlen (optional, falls nur diese benÃ¶tigt werden)
result = grouped_station1[[f'{col}_int' for col in columns_to_process]]

result.head()
```

```python
fig = px.line(result)
fig
```

```python
columns_to_average = ['value_1', 'value_2', 'value_3', "value_4", "value_5"]
landpm25['datum'] = pd.to_datetime(landpm25['createdAt'])
avg_landpm10 = landpm10.groupby('datum')[columns_to_average].mean()
avg_landpm25 = landpm25.groupby('datum')[columns_to_average].mean()
avg_landpm10 = avg_landpm10.reset_index()
avg_landpm25 = avg_landpm25.reset_index()
avg_landpm10['average_value'] = avg_landpm10[columns_to_average].mean(axis=1)
avg_landpm25['average_value'] = avg_landpm25[columns_to_average].mean(axis=1)
avg_landpm10 = avg_landpm10.dropna(subset="average_value")
avg_landpm25 = avg_landpm25.dropna(subset="average_value")
avg_landpm25
```

```python
fig = px.line(avg_landpm25, x="datum", y="average_value")
fig
```
