```python
import pandas as pd
import plotly.express as px
```

```python
landpm25 = pd.read_csv("werte_daten/werte_land_PM2_5.csv")
landpm10 = pd.read_csv("werte_daten/werte_land_PM10.csv")

stadtpm25 = pd.read_csv("werte_daten/werte_stadt_PM2_5.csv")
stadtpm10 = pd.read_csv("werte_daten/werte_stadt_PM10.csv")

stadtpm10

```

```python
landpm10['datum'] = pd.to_datetime(landpm10['createdAt'])
stadtpm10['datum'] = pd.to_datetime(stadtpm10['createdAt'])
columns_to_process = ['value_1', 'value_2', 'value_3', 'value_4', 'value_5']

# Konvertiere die Spalten zu float, falls sie nicht bereits numerisch sind
stadtpm10[columns_to_process] = stadtpm10[columns_to_process].apply(pd.to_numeric)

# Interpolieren und dann gruppieren
for col in columns_to_process:
    stadtpm10[f'{col}_int'] = stadtpm10[col].interpolate()

# Gruppieren nach 'datum' und Berechnung des Mittelwerts für interpolierte Spalten
# Hier müssen wir sicherstellen, dass wir die numerischen Typen erhalten
grouped_station1 = stadtpm10.groupby('datum').mean(numeric_only=True)

# Nur die interpolierten Spalten auswählen (optional, falls nur diese benötigt werden)
result = grouped_station1[[f'{col}_int' for col in columns_to_process]]

result.head()
```

```python
fig = px.line(result)
fig
```

```python
columns_to_average = ['value_1', 'value_2', 'value_3', "value_4", "value_5"]
landpm25['datum'] = pd.to_datetime(landpm25['createdAt'])
stadtpm10['datum'] = pd.to_datetime(stadtpm10['createdAt'])
stadtpm25['datum'] = pd.to_datetime(stadtpm25['createdAt'])
avg_landpm10 = landpm10.groupby('datum')[columns_to_average].mean()
avg_landpm25 = landpm25.groupby('datum')[columns_to_average].mean()
avg_stadtpm10 = stadtpm10.groupby('datum')[columns_to_average].mean()
avg_stadtpm25 = stadtpm25.groupby('datum')[columns_to_average].mean()
avg_landpm10 = avg_landpm10.reset_index()
avg_landpm25 = avg_landpm25.reset_index()
avg_stadtpm10 = avg_stadtpm10.reset_index()
avg_stadtpm25 = avg_stadtpm25.reset_index()
avg_landpm10['average_landpm10'] = avg_landpm10[columns_to_average].mean(axis=1)
avg_landpm25['average_landpm2_5'] = avg_landpm25[columns_to_average].mean(axis=1)
avg_stadtpm10['average_stadtpm10'] = avg_stadtpm10[columns_to_average].mean(axis=1)
avg_stadtpm25['average_stadtpm2_5'] = avg_stadtpm25[columns_to_average].mean(axis=1)
avg_landpm10 = avg_landpm10.dropna(subset="average_landpm10")
avg_landpm25 = avg_landpm25.dropna(subset="average_landpm2_5")
avg_stadtpm10 = avg_stadtpm10.dropna(subset="average_stadtpm10")
avg_stadtpm25 = avg_stadtpm25.dropna(subset="average_stadtpm2_5")
avg_stadtpm25
```

```python
fig = px.line(avg_stadtpm25, x="datum", y="average_stadtpm2_5")
fig
```

```python
average_pm10 = pd.merge(avg_landpm10, avg_stadtpm10, on="datum")
average_2_5 = pd.merge(avg_landpm25, avg_stadtpm25, on="datum")

average_values_raw = pd.merge(average_2_5, average_pm10, on="datum")
average_values = average_values_raw[["datum", "average_landpm10", "average_landpm2_5", "average_stadtpm10", "average_stadtpm2_5"]]
average_values
```

```python
fig = px.line(average_values, x="datum", y=["average_landpm10", "average_landpm2_5", "average_stadtpm10", "average_stadtpm2_5"])
fig
```

```python
stadt = px.line(average_values, x="datum", y=["average_stadtpm10", "average_stadtpm2_5"], title="Luftqualität Stadt")
stadt
```
