```python
import pandas as pd
import requests
import geopandas as gpd
import csv
import plotly.express as px
```

```python
url = "https://api.opensensemap.org/boxes/5b267cac1fef04001b78f88b"
daten = requests.get(url).json()
daten
```

```python
url = "https://api.opensensemap.org/boxes/5b267cac1fef04001b78f88b/data/5b267cac1fef04001b78f890?from-date=2024-05-24T00:00:00.000Z&to-date=2024-05-24T20:44:01.434Z&download=true&format=json"
daten = requests.get(url).json()
df = pd.DataFrame(daten)
df['createdAt'] = pd.to_datetime(df['createdAt'])
df["value"] = pd.to_numeric(df['value'])
df
```

```python
fig = px.bar(df, x="createdAt", y="value", color_continuous_scale="Reds")
fig
```
